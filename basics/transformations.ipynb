Firstly, let's define and plot the square determined by its four vertices. Why do we use three coordinates?

```python
quad =  np.array([
            [0.,0.,1],
            [0.,1.,1],
            [1.,1.,1],
            [1.,0.,1]
        ]).T  
```

```python
display_vertices(quad, 'Original Quad')
```

we rotate our sqare by 45 degrees and plot it.

```python
rotated_quad = transform_vertices(quad, mrotate(45))
display_vertices(rotated_quad, 'Rotated Quad')
```

Once everything is implemented, it's time to investigate the four transformations t1,t2,t3,t4.

```python
t1, t2, t3, t4 = define_transformations()
```

```python
image1_vertices = transform_vertices(quad, t1)
display_vertices(image1_vertices, 'Target Image 1')
```

```python
image2_vertices = transform_vertices(quad, t2)
display_vertices(image2_vertices, 'Target Image 2')
```

```python
image3_vertices = transform_vertices(quad, t3)
display_vertices(image3_vertices, 'Target Image 3')
```

```python
image4_vertices = transform_vertices(quad, t4)
display_vertices(image4_vertices, 'Target Image 4')
```
